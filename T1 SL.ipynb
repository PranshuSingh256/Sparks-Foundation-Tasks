### 1 - Importing the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl
%matplotlib inline
import seaborn as sns

import warnings as wg
wg.filterwarnings("ignore")
```

```python
data = pd.read_csv("http://bit.ly/w-data") #C:/Users/Administrator/Documents/Jupyter Notebook/T1_data.csv
print("Data Import Completed")
print("Printing first 5 data-set:")
data.head()
#data.tail(3)
```

```python
#data.shape
#data.describe()
#data.isnull().sum()
#data.columns
data.info()
```

### 2 - Visualizing the dataset

```python
mpl.rcParams["figure.figsize"] = [7,4]
data.plot(x='Hours', y='Scores', style = 'x', color='blue', markersize='5')
mpl.title("HOURS VS PERCENTAGE")
mpl.xlabel("Total Hours Studied",size='8')
mpl.ylabel("Percentage Scored",size='8')
mpl.grid()
mpl.show()
```

### 3 - Preparing Data

```python
x = data.iloc[:,:-1].values     #vertical row vector & last column of data frame
y = data.iloc[:,1].values       #horizontal row vector & 2nd column of data frame

# we can also use .corr to determine the corelation between the variables 
data.corr()
```

### 4 - Training of Model

```python
#80% of data is taken as training data

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
reg = LinearRegression()
reg.fit(x_train, y_train)
```

### 5 - Visualizing the model

```python
line_plot = reg.coef_ * x + reg.intercept_
mpl.title("Line Regression VS Trained Model")
mpl.rcParams["figure.figsize"] = [7,5]
mpl.scatter(x, y, color = 'red')
mpl.xlabel("Total Hours Studied",size='8')
mpl.ylabel("Percentage Scored",size='8')
mpl.plot(x, line_plot, markersize = '4')
mpl.grid()
mpl.show()

#sns.regplot(x=data['Hours'], y=data['Scores'])
```

### 6 - Making Predictions

```python
print(x_test)
print("Predicting the Scores:")
y_predict = reg.predict(x_test)
print(y_predict)
```

```python
dataf = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_predict})
mpl.rcParams["figure.figsize"] = [5.5,3]
dataf.plot(kind = 'bar')
```

```python
mpl.grid(True)
sns.kdeplot(y_predict, label="Predicted", shade=True)
sns.kdeplot(data=y_test, label="Actual", shade=True)
```

```python
#Predicting the Score if a student studies for 9.25 hrs/day
hours = [[9.25]]
predict = reg.predict(hours)
print(predict)
```

### 7 - Evaluating the model

```python
from sklearn import metrics
print('Mean Ambsolute Error:', metrics.mean_absolute_error(y_test, y_predict))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,y_predict))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_absolute_error(y_test, y_predict)))
print('Slope of Regression Line', reg.coef_)
print('Slope of Regression Line', reg.intercept_)
```
